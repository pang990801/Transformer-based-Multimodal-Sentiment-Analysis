```python
import pickle
import matplotlib.pyplot as plt
# 定义文件路径
pkl_file_path = 'MOSEI/mosei_raw_bert_old.pkl'


# 打开并加载 pkl 文件
with open(pkl_file_path, 'rb') as file:
    data = pickle.load(file)

# 打印加载的数据
# print(data['train']['id'])
# print(data['test']['id'])
print(data['valid']['text'][0].shape)   # 查看数据内容（根据需要适当格式化输出）

def print_structure(data, indent=0):
    """递归打印数据结构"""
    prefix = " " * indent
    if isinstance(data, dict):
        print(f"{prefix}Dictionary with {len(data)} keys:")
        for key, value in data.items():
            print(f"{prefix}- Key: {key}")
            print_structure(value, indent + 4)
    elif isinstance(data, list):
        print(f"{prefix}List with {len(data)} elements:")
        if len(data) > 0:
            print_structure(data[0], indent + 4)  # 递归打印列表的第一个元素
    elif isinstance(data, tuple):
        print(f"{prefix}Tuple with {len(data)} elements:")
        for i, value in enumerate(data):
            print(f"{prefix}- Element {i}:")
            print_structure(value, indent + 4)
    else:
        print(f"{prefix}Value of type {type(data)}: {str(data)[:100]}")  # 打印前100字符以避免过长

print_structure(data)

# print(data['valid']['vision'][0][:30])
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the pkl file (assuming the user will upload it)
file_path = "MOSEI/mosei_raw_bert.pkl"  # Replace with the actual file path
with open(file_path, 'rb') as file:
    data = pickle.load(file)

# Extract labels from train, valid, and test datasets
train_labels = data['train']['labels'][:, 0, 0]
valid_labels = data['valid']['labels'][:, 0, 0]
test_labels = data['test']['labels'][:, 0, 0]

# Combine all labels
all_labels = np.concatenate([train_labels, valid_labels, test_labels])

# Set Seaborn style and academic color palette
sns.set_theme(style="whitegrid")
sns.set_palette("muted")  # Academic-style muted color palette

# Create the plot
plt.figure(figsize=(12, 6))  # Set figure size
ax = sns.histplot(
    all_labels,
    bins=15,  # Fewer bins for better readability
    kde=True,  # Add kernel density estimation
    color=sns.color_palette("muted")[0],  # Use the first color from muted palette
    edgecolor="black",
    alpha=0.8,
    label="Sentiment Distribution"
)

# Modify the legend, axis labels, and title
plt.title("CMU-MOSEI Sentiment Label Distribution", fontsize=22, weight="bold")
plt.xlabel("Sentiment Intensity [-3, 3]", fontsize=20)
plt.ylabel("Number of Samples", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.tight_layout()  # Automatically adjust layout
plt.show()
```
